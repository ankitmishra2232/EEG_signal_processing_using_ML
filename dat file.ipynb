```python
import os
import numpy as np

# Set the path to the directory containing the .dat files
dir_path = "E:/Signal/data_preprocessed_python/"

# Get a list of all the .dat files in the directory
dat_files = [file for file in os.listdir(dir_path) if file.endswith(".dat")]
raws = []
# Loop through each file and read the data
for dat_file in dat_files:
    # Construct the full path to the file
    file_path = os.path.join(dir_path, dat_file)
    
    # Read the data from the file
    data = np.fromfile(file_path, dtype=np.float32)
    raws.append(data)
```
